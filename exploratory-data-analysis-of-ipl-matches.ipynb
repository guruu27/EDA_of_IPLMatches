```python
#extracting the records where team that won did batting first
batting_first = df[df['win_by_runs']!=0]
batting_first
```

```python
#extracting the records where team that won did fielding first
field_first = df[df['win_by_runs']==0]
field_first
```

```python
#making a plot for distribution of runs
plt.figure(figsize=(6,5))
plt.hist(batting_first['win_by_runs'])
plt.title('distribution of runs')
plt.show()
```

```python
#number of wins wrt to teams who batted first
wins = batting_first['winner'].value_counts()
wins
```

```python
#plotting with top 3 teams who batted first with most number of wins

plt.figure(figsize=(7,7))
plt.bar(batting_first['winner'].value_counts()[0:3].keys(),batting_first['winner'].value_counts()[0:3],color=['orange','blue','green'])
plt.xlabel('Team Names')
plt.ylabel('Frequency')
plt.xticks(rotation=40);
```

```python
plt.figure(figsize=(6,6))
plt.pie(batting_first['winner'].value_counts(),labels=batting_first['winner'].value_counts().keys(),shadow=True,autopct='%0.1f%%')
plt.axis('equal');
```

```python
## Has Toss-winning helped in Match-winning?
# find out how many times a team has won the match after winning the toss
import numpy as np
filter = np.sum(df['toss_winner']== df['winner'])
rows = df.index
percentage = filter / len(rows) *100
print(f'number of times a team has won the match after winning toss is {filter}')
print(f'percent of  a team has won the match after winning toss is {percentage}%' )
if percentage>= 70:
    print('toss winning helped in match winning')
else:
    print('no, toss winning did not helped in match winning')
```
